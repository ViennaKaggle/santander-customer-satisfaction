```python
%matplotlib inline
import skutils
import pandas as pd

TRAIN_DATA = "../../data/train.csv"
```

```python
data = pd.read_csv(TRAIN_DATA)
```

```python
def transform(data):
    
    # Extract the ids
    ids = data['ID']
    
    # Extract the data
    X = data.drop(['TARGET', 'ID'],  axis=1)
    
    # Extract the labels
    y = data['TARGET']
    
    return ids, X, y
```

```python
# Transform data
ids, X, y = transform(data)

# Missing Values
NanHandler = skutils.NanPreProcessor().fit(X, y)

# Fill Nan with Interclass Median
X = NanHandler.transform(X)
```

```python
X.describe()
```

```python
skutils.tsne_plot(X.values, y.values, n_samples=5000)
```
