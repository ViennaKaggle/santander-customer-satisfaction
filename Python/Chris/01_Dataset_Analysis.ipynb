```python
%matplotlib inline
import skutils
import pandas as pd

TRAIN_DATA = "../../data/train.csv"
TEST_DATA = "../../data/train.csv"
```

```python
# Load the training data
train = pd.read_csv(TRAIN_DATA)

# Load the testing data
test = pd.read_csv(TRAIN_DATA)
```

```python
# Print pretty stats about the training data set
skutils.pretty_stats(train)
```

```python
# Print pretty stats about the testing data set
skutils.pretty_stats(test)
```

```python
def transform(data):
    
    # Extract the ids
    ids = data['ID']
    
    # Extract the data
    X = data.drop(['TARGET', 'ID'],  axis=1)
    
    # Extract the labels
    y = data['TARGET']
    
    return ids, X, y
```

```python
# Transform data
ids, X, y = transform(train)
```

```python
skutils.tsne_plot(X.values, y.values, n_samples=5000)
```
